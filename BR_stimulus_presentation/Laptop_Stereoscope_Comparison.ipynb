for each row in laptopDf (uncorrected)
- the 0th item is 1
- the 1th item is 'JH' (participant)
- 2ns is trial number
- 3rd is top color
- 4th is top orientation 
- 5th is 0.5??? likely red opacity
- 6th is 1.0??? likely run 1 (of an experiment)
- 7th seems to reflect trial order 
- 8th starts with KeyDur - string (subsequent items in the array are floats

```python
# correct the format of laptopDf
laptopDf_c = pd.DataFrame(columns=['topOri', 'topColor','keyDur', 'keyRT', 'keyName'])


for i, row in laptopDf.iterrows():
    responseLength = 0
    keyDur = []
    
    keyDurStart = i[8]
    keyDur.append(float(keyDurStart[1:]))
    
    foundResponse = 0
    responseIndex = 9
    
    
    #print(keyDurStart, i[responseIndex])
    
    while ~foundResponse:

        # check to see if the next item in the dataframe is a float or a string
        if isinstance(i[responseIndex], str):
            # check to see if it ends with a bracket
            if i[responseIndex][-1] ==']':
                try:
                    #print(len(keyDur))
                    #print('foundResponse!')
                    foundResponse = 1
                    keyDur.append(float(i[responseIndex][:-1]))
                    break
                except:
                    print('unexpected value: ', i[responseIndex])
            else:
                keyDur.append(float(i[responseIndex]))
                
        elif isinstance(i[responseIndex], float):
            keyDur.append(i[responseIndex])
            
        responseIndex += 1
        

    
    responseLength = len(keyDur)
#     print(responseLength)
#     print(i[responseLength+responseIndex])
    responseIndex += 1
    
    keyRT = []
    
    # start adding keyRT
    if isinstance(i[responseIndex], str):
        if i[responseIndex][0] =='[':
            keyRT.append(float(i[responseIndex][1:]))
            #print('RT start:', i[responseIndex])
        else:
            print('ERROR!', i[responseIndex])
        
    
    count = 0
    for index in range(responseIndex, responseIndex+responseLength):
        count+= 1
        
        if isinstance(i[index], str):
            # check to see if it ends with a bracket
            #print(i[index])
            if i[index][-1] ==']':
                try:
                   
                    keyRT.append(float(i[index][:-1]))
                    #print('RT end: ', i[index])
                   
                except:
                    print('unexpected value: ', i[index])
            else:
                if i[index][0] is not '[':
                    #print(i[index][0])
                    keyRT.append(float(i[index]))
        elif isinstance(i[index], float):
            keyRT.append(i[index])

        else:
            print(i[index], '!!!')
            
        
    print(len(keyDur), len(keyRT), count)
    responseIndex += responseLength 
    #responseIndex += 1
    
    keyName = []
    
    if isinstance(i[responseIndex], str):
        if i[responseIndex][0] =='[':
            keyName.append(i[responseIndex][2:-1])
            print('Name start: ', i[responseIndex])
        else:
            print('ERROR - keyName')
        
    
    for index in range(responseIndex+1, responseIndex+responseLength):
        if isinstance(i[index], str):
            # check to see if it ends with a bracket
            if i[index][-1] ==']':
                try:

                   
                    keyName.append(i[index][2:-2])
                    print('name end: ', i[index])
                except:
                    print('unexpected value: ', i[index])
            else:
                keyName.append(i[index][2:-1]) # indices were found through trial and error 
            
    
    
#     print(len(keyDur), len(keyRT), len(keyName))
    
    laptopDf_c.loc[len(laptopDf_c.index)] = [i[4], i[3], keyDur, keyRT, keyName]
        
    
    
```

```python
redDurations = []
blueDurations = []
mixedDurations = []

leftDurations = [] # 135
rightDurations = [] # 45
```

```python
for i, row in laptopDf_c.iterrows():
    responseLen = len(row['keyDur'])
    
    blueIs45 = 0
    
    if row['topColor'] == 'blue':
        if row['topOri'] == 45:
            blueIs45 = 1
        else:
            blueIs45 = 0
    else:
        if row['topOri'] == 45: # red is 45
            blueIs45 = 0
        else:
            blueIs45 = 1
    
    for j in range(responseLen):
        if row['keyName'][j] == 'right': # blue
            blueDurations.append(row['keyDur'][j])
            
            if blueIs45: # if true blue is 45 
                rightDurations.append(row['keyDur'][j])
            else: # blue is not 45
                leftDurations.append(row['keyDur'][j])
            
        elif row['keyName'][j] == 'left': # red
            redDurations.append(row['keyDur'][j])
            
            if blueIs45: # if true blue is 45 and red is 135 (left)
                leftDurations.append(row['keyDur'][j])
            else: # blue is not 45
                rightDurations.append(row['keyDur'][j])
        elif row['keyName'][j] == 'space': # mixed
            mixedDurations.append(row['keyDur'][j])
        
        
```
